```python
import cv2

import numpy as np

from matplotlib import pyplot as plt
```

```python
file_name = 'P0124'
image_path = '/Users/usmanr/workspace/rotatedBoxes/training/images/1/images/{0}.png'.format(file_name)

# label_path = '/Users/usmanr/workspace/rotatedBoxes/training/labelTxt-v1.0/trainset_reclabelTxt/P0005.txt'

label_path = '/Users/usmanr/workspace/rotatedBoxes/training/labelTxt-v1.0/trainset_polygon/labelTxt-v1.5/{0}.txt'.format(file_name)
```

```python
with open(label_path) as f:
    lines = f.readlines()
```

```python
lines = lines[2:]
```

```python
labels = []
for line in lines:
    l = line.strip().split(' ')
    x = list(map(float, l[:8]))
    x = list(map(int, x))
    labels.append( x + l[8:] )
    print(list(map(float, l[:8])))
```

```python
labels
```

```python
# Blue color in BGR
color = (255, 0, 0)
 
# Line thickness of 2 px
thickness = 1
```

```python
image = cv2.imread(image_path)
```

```python
for label in labels:
    
    polygon = np.array(label[:8]).reshape(4, 2)
    
    polygon = polygon.reshape((-1, 1, 2))

    image = cv2.polylines(image, [polygon],
                          True, color, 2)
    
plt.figure(figsize=(15,15))
plt.imshow(image)
```
